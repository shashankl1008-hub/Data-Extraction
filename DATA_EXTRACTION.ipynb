Ans Q2---

Data extraction is the process of collecting data from various source systems such as databases, files, APIs, and cloud applications. It is the first stage of the ETL pipeline, where data is retrieved and prepared for transformation and loading into a target system like a data warehouse. Data extraction enables organizations to consolidate data from multiple sources for reporting, analytics, and decision-making.

Ans Q4--

The steps involved in extracting data from a relational database are:

Connect to the database.
Identify the required tables and columns.
Write SQL queries to retrieve data.
Apply filters if needed.
Perform incremental extraction for efficiency.
Validate the extracted data.
Load the data into a staging area.
Log and monitor the extraction process.

Example: Extracting customer data from a MySQL database using a SQL query and loading it into a staging area before transforming and loading it into a data warehouse.

Ans Q 5--

Three common challenges during data extraction are:

Poor Data Quality – Source data may contain missing, duplicate, or incorrect records.
Data Format Differences – Data from databases, files, and APIs often have different structures and formats.
Large Data Volumes – Extracting large amounts of data can affect performance and increase processing time.

These challenges are typically addressed through data validation, standardization, incremental extraction, and query optimization.

An API (Application Programming Interface) is a mechanism that allows different software applications to communicate and exchange data. In ETL, APIs are used to extract data from applications such as Shopify, Salesforce, and Google Analytics. APIs support real-time or near real-time data extraction by automatically providing the latest data whenever records are created or updated, enabling faster reporting and analytics.

Ans Q 7--


Databases are preferred for enterprise-level data extraction because they efficiently store and manage large volumes of structured data, support fast SQL-based querying, ensure data integrity and security, and enable incremental extraction. They also integrate easily with ETL tools and provide high reliability, making them ideal for large-scale business reporting and analytics.

Ans Q 8--

When extracting data from large CSV files (1GB+), an ETL developer should analyze the file structure, read data in chunks instead of loading it entirely into memory, extract only required columns, validate data quality, use staging tables, implement incremental processing, monitor extraction logs, and optimize performance through compression and parallel processing. These practices improve efficiency, scalability, and reliability of the ETL process.
